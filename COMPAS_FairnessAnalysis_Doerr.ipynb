## 1. Einführung


Bias in der Künstlichen Intelligenz bezeichnet systematische Verzerrungen in Daten, Modellen oder ihrem Einsatz, die dazu führen, dass bestimmte Personen oder Gruppen (z. B. nach Geschlecht, Ethnie, Alter) schlechter oder besser behandelt werden als andere – unabhängig davon, ob diese Eigenschaften explizit im Modell verwendet werden oder nicht.  
Bias kann in **allen** Phasen der ML-Pipeline entstehen
- **Datenerfassung & -aufbereitung**  
- **Modelltraining & -auswahl**  
- **Evaluation & Deployment**

Auch wenn wir uns größte Mühe geben, **vollkommen faire** Modelle zu bauen, ist Bias in KI-Systemen **grundsätzlich unvermeidlich**

## 2. Metakategorischer Rahmen (MDPI)



Ein aktueller MDPI-Review unterscheidet Bias-Quellen nach ihrer **Ursprungsebene** :

1. **Data Bias**  
   Verzerrungen, die direkt aus den Daten resultieren (Sampling Bias, Measurement Bias, Label Bias).  
2. **Algorithmic Bias**  
   Verzerrungen, die durch Modellarchitektur, Trainingsprozess oder Optimierungsziele eingeführt werden (Aggregation Bias, Feedback-Loops).  
3. **User Bias**  
   Verzerrungen, die aus der Interaktion von Menschen mit dem System oder durch den Einsatz-Kontext entstehen (Deployment Bias, Automation Bias
   
   
## 3. Sechs Bias-Typen


Um diese Ursprungsebenen nch genauer definieren zu können orientieren wir uns im Folgenden an Alexis Cooks Tutorial auf Kaggle (Cooks, 2020).
### 3.1 Historical Bias  
**Definition:** Historische Verzerrung liegt vor, wenn Trainingsdaten die (verzerrte) Realität vergangener Entscheidungen abbilden.  
- **Phase:** Datenerfassung  
- **Ursache:** Gesellschaftlich gewachsene Ungleichheiten und strukturelle Diskriminierung in historischen Daten  
- **Beispiel (Redlining):** Zwischen den 1930er Jahren und heute wurden Minority-Viertel in den USA durch „Redlining“ systematisch von Hypothekendarlehen ausgeschlossen – ein Effekt, der bis heute in Bonitätsmodellen fortwirkt.

### 3.2 Representation Bias  
**Definition:** Repräsentationsverzerrung entsteht, wenn die Stichprobe die Zielpopulation nicht angemessen abbildet.  
- **Phase:** Sampling (Datenerfassung)  
- **Ursache:** Unausgewogenes Stichproben-Design, das bestimmte Gruppen unter- oder überrepräsentiert  
- **Beispiel (Gender Shades):** Dunkelhäutige Frauen wurden von kommerziellen Gesichts­erkennungssystemen bis zu 34,7 % falsch klassifiziert, während hellhäutige Männer nur 0,8 % Fehlerrate hatten.

### 3.3 Measurement Bias  
**Definition:** Ungleiche Messgenauigkeit verzerrt Features/Labels.  
- **Phase:** Featurisierung & Labeling  
- **Ursache:** Verwendung von Proxy-Variablen, deren Qualität oder Repräsentativität zwischen den Gruppen variiert  
- **Beispiel (Gesundheitskosten):** Ein Algorithmus, der hohe Gesundheitskosten als Proxy für Krankheitsrisiko nutzt, identifizierte nur halb so viele schwarze wie weiße Hochrisikopatient*innen, weil schwarze Patient*innen seltener behandelt wurden und somit geringere Kosten verursachten.

### 3.4 Aggregation Bias  
**Definition:** Aggregationsverzerrung liegt vor, wenn heterogene Subgruppen in einem Modell zusammengefasst werden, ohne ihre spezifischen Merkmale zu berücksichtigen.  
- **Phase:** Modelltraining  
- **Ursache:** Ignorieren subgroup-spezifischer Verteilungen und Merkmale bei einem einheitlichen Modell  
- **Beispiel (Diabetes-Modelle):** Hispanic-Patient*innen hatten 2022 eine 60 % höhere diagnostizierte Diabetesrate als nicht-Hispanic-Weiße. Ein generisches Modell ohne Berücksichtigung von Ethnizität liefert für diese Gruppe ungenaue Vorhersagen (US Gov zit. nach Cooks, 2025).

### 3.5 Evaluation Bias  
**Definition:** Evaluationsverzerrung entsteht, wenn Benchmark-Datensätze nicht die tatsächliche Anwenderpopulation widerspiegeln.  
- **Phase:** Testing & Benchmarking  
- **Ursache:** Nicht repräsentative Testdatensätze bzw. Benchmarks  
- **Beispiel (IJB-A & Adience):** Die häufig genutzten Datensätze enthalten zu 79,6 % bzw. 86,2 % hellhäutige Gesichter. Modelle, die auf diesen Benchmarks trainiert oder evaluiert werden, wirken auf diverseren Gruppen oft deutlich schlechter.

### 3.6 Deployment Bias  
**Definition:** Deployment-Bias tritt auf, wenn ein Modell in der Praxis anders verwendet wird, als ursprünglich konzipiert.  
- **Phase:** Einsatz & Anwendung  
- **Ursache:** Missbrauch oder Übernutzung des Modells außerhalb seines intendierten
- **Beispiel (PRECOBS):** PRECOBS ist eine prädiktive Polizeisoftware, die in Bayern seit 2014 (u. a. in München und Nürnberg) genutzt wird, um „Near-Repeat“-Einbrüche vorherzusagen. Ursprünglich als Planungs- und Einsatzhilfsmittel gedacht, wird sie in der Praxis oft zur gezielten Straßenziehung und Ressourcenumverteilung herangezogen. Die enge Fokussierung auf vermeintliche „Hotspots“ ohne Ergänzung durch lokale Expertise zeigt, wie das Modell über seine beabsichtigte Rolle hinaus eingesetzt wird (John Jay College of Criminal Justice, 2024 & SozTheo, o.J.).
- **Beispiel (Skala-RiskTool):** In einigen deutschen Gerichten und Bewährungseinrichtungen kommt die „Skala“ zum Einsatz – ein instrumentiertes Prognosemodell zur Rückfallwahrscheinlichkeit. Obwohl Skala nur als beratendes Hilfsmittel vorgesehen ist, nutzen Richter:innen und Bewährungshelfer:innen die Scores häufig als bindende Entscheidungsgrundlage für Bewährungsauflagen oder vorzeitige Haftentlassungen. Diese Diskrepanz zwischen geplanter Rolle und tatsächlicher Anwendung ist ein klassischer Fall von Deployment Bias (SozTheo, o.J.).

## 4. Hands-On: Bias-Analyse mit dem COMPAS-Datensatz







### Datensatz-Überblick: COMPAS Recidivism Racial Bias

**Herkunft & Kontext**  
- Quelle: ProPublica / Kaggle (Dataset “danofer/compass”)  
- Beobachtungszeitraum: 2013–2014 in Broward County, Florida  
- Zweck: Analyse, ob der COMPAS‐Score Rückfallrisiken verzerrt vorhersagt  

**Kurzbeschreibung**

Der COMPAS-Datensatz umfasst 6 172 Fälle mit Informationen zu Rückfall innerhalb von 2 Jahren (`Two_yr_Recidivism`), Vorstrafenanzahl, COMPAS-Score sowie demografischen Merkmalen (Alter, ethnische Zugehörigkeit, Geschlecht) und Delikt-Schwere aus dem Broward County zwischen den Jahren 2013 und 2014.

**Analyseziel**

Wir nutzen ihn, um zu untersuchen, ob das Modell systematisch unterschiedliche Vorhersagen und Fehlerraten für Schwarze vs. Nicht-Schwarze liefert (Group Fairness). Konkret messen wir Demographic Parity Difference und False-Positive-Rate Difference, um Bias-Typen wie Historical und Measurement Bias sichtbar zu machen.  

### 4.1. Datensatz laden

```python

import pandas as pd

# Wir laden die bereits vorverarbeitete CSV-Datei mit COMPAS-Fällen
# (One-Hot encodierte Alters- und Race-Kategorien).
df = pd.read_csv('Compas Data.csv')

# Kontrolle: Anzahl Zeilen/Spalten und Spaltennamen ausgeben
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())

df.head()  # Erster Blick auf die Daten
```

### Datensatz-Überblick: COMPAS Recidivism Racial Bias

**Inhalt & Schlüsselmerkmale**

| `score_factor`           | COMPAS-Score (1 = hohes Risiko, 0 = niedriges Risiko)                                         |

Mit diesen Einblicken haben wir die Grundlagen gelegt, um im nächsten Schritt gezielt Bias-Kennzahlen zu berechnen und zu visualisieren.  

### 4.2 Vorverarbeitung

```python
# 1) Zielvariable setzen: Rückfall innerhalb von 2 Jahren
df['y_true'] = df['Two_yr_Recidivism']

# 2) Sensibles Gruppenmerkmal: 'is_black' direkt aus der 'African_American' Spalte in neue Spalte
#    'is_black' kopiert.
#    1 = African_American, 0 = alle anderen
df['is_black'] = df['African_American']

# 3) Auswahl der Features:
#    - Number_of_Priors: Anzahl vorheriger Delikte
#    - score_factor: COMPAS-Score (binär)
#    - is_black: sensibles Merkmal für Gruppenvergleich
features = ['Number_of_Priors', 'score_factor', 'is_black']
X = df[features]
y = df['y_true']

# Kontrolle: Verteilt sich das sensitive Merkmal wie erwartet?
# Integer Werte von 0 bis 1 (int64 - 64 Bit Zahl) indizieren non-Black und Black aus der Spalte 'is_black'.
print("Black vs. non-Black counts:\n", df['is_black'].value_counts())
```

### 4.4 Modelltraining

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split


# 1) Logistic Regression:
# die Koeffizienten zeigen direkt, wie sich jede Variable (inkl. des sensiblen Attributes) auf die Klassifikation auswirkt
# schnell & ressourcenschonend trainiert, ideal für kleine bis mittelgroße Datensätze wie COMPAS

# 2) Stratified Train/Test-Split:
#    stratify=df['is_black'] #    Wir halten das Verhältnis von Schwarzen vs. Nicht-Schwarzen in beiden Teilmengen konstant 
#    um sicherzustellen, dass unser Modelltraining und unsere Auswertung nicht durch
#    ungleiche Gruppenverteilungen verfälscht werden
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    stratify=df['is_black'],
    test_size=0.3,
    random_state=42
)

# 2) Modell initialisieren & trainieren:
model = LogisticRegression(solver='liblinear')
model.fit(X_train, y_train)

# 3) Vorhersagen erzeugen und zusammenführen:
df_test = X_test.copy()
df_test['y_true']  = y_test.values           # wahre Labels
df_test['y_pred']  = model.predict(X_test)   # Modell-Vorhersagen
df_test['is_black'] = X_test['is_black'].values  # Gruppe erneut anhängen

df_test.head()  # Kontrolle der neuen Spalten
```

Interpretation

| Index | Number_of_Priors | score_factor | is_black | y_true | y_pred | Bedeutung                                  |
|-------|------------------|--------------|----------|--------|--------|---------------------------------------------|
| 51    | 5                | 0            | 1        | 0      | 0      | Eine schwarze Person mit 5 Vorstrafen, COMPAS-Score niedrig, hatte wirklich keinen Rückfall und unser Modell sagt korrekt „kein Rückfall“ (True Negative). |
| 111   | 2                | 0            | 0        | 0      | 0      | Eine nicht-schwarze Person mit 2 Vorstrafen, COMPAS-Score niedrig, kein Rückfall – Modell trifft erneut korrekt (TN). |
| 4078  | 2                | 1            | 0        | 0      | 1      | Eine nicht-schwarze Person mit 2 Vorstrafen, Score hoch, hatte **keinen** Rückfall, Modell sagt fälschlich „Rückfall“ (False Positive). |
| 3194  | 2                | 1            | 0        | 1      | 1      | Nicht-schwarze Person, Score hoch, hatte Rückfall, Modell trifft korrekt „Rückfall“ (True Positive). |
| 4210  | 0                | 1            | 1        | 1      | 1      | Schwarze Person ohne Vorstrafen, Score hoch, hatte Rückfall, Modell trifft korrekt „Rückfall“ (TP). |

**Spalten im Detail:**
- `Number_of_Priors`: Wie viele frühere Straftaten registriert sind.
- `score_factor`: COMPAS-Score (0 = niedriges Risiko, 1 = hohes Risiko).
- `is_black`: 1 = Person ist als „African American“ kodiert, 0 = sonst.
- `y_true`: Tatsächliches Auftreten eines Rückfalls (0 = nein, 1 = ja).
- `y_pred`: Vom Modell vorhergesagt (0 = Modell sagt „kein Rückfall“, 1 = Modell sagt „Rückfall“).

**Was lernen wir daraus?**
- **True Negatives (TN):** Modell erkennt „kein Rückfall“ korrekt (z. B. Index 51, 111).  
- **False Positives (FP):** Modell sagt „Rückfall“, obwohl keiner stattfindet (z. B. Index 4078) – hier zahlen wir eine Gruppe fälschlich als riskant.  
- **True Positives (TP):** Modell sagt „Rückfall“ und das tritt ein (Index 3194, 4210).  
- **False Negatives (FN):** (Nicht gezeigt) wären Fälle, in denen `y_true = 1` aber `y_pred = 0`.

Mithilfe dieser ersten Kontrolle sehen wir schon, welche Arten von Fehlklassifikationen (FP vs. FN) auftreten und können sie später gruppenweise (is_black = 0 vs. 1) quantifizieren, um Bias besser zu verstehen.  

### 4.5 Fairness - Metriken berechnen

```python
from sklearn.metrics import confusion_matrix

def demographic_parity_diff(df, group_col, y_pred_col):
    """
    Berechnet die absolute Differenz der positiven Vorhersageraten
    (Demographic Parity) zwischen den Gruppen.
    """
    # 1) Gruppiere das DataFrame nach dem sensiblen Merkmal und berechne
    #    für jede Gruppe die durchschnittliche positive Vorhersagerate.
    rates = df.groupby(group_col)[y_pred_col].mean()
    # 2) Ziehe die beiden Raten voneinander ab und gib den absoluten Wert zurück.
    return abs(rates.diff().iloc[-1])

def false_positive_rate_diff(df, group_col, y_true_col, y_pred_col):
    """
    Berechnet die absolute Differenz der False-Positive-Raten
    zwischen den Gruppen.
    """
    fprs = []  # Liste zum Sammeln der FP-Raten pro Gruppe
    # 1) Iteriere über jede Gruppe im DataFrame
    for g, sub in df.groupby(group_col):
        # 2a) Erstelle eine Konfusionsmatrix (TN, FP, FN, TP)
        tn, fp, fn, tp = confusion_matrix(sub[y_true_col], sub[y_pred_col]).ravel()
        # 2b) Berechne die False-Positive-Rate = FP / (FP + TN)
        fpr = fp / (fp + tn)
        fprs.append(fpr)  # Füge die FPR-Liste hinzu
    # 3) Differenz zwischen der höchsten und niedrigsten FPR
    return abs(max(fprs) - min(fprs))

# 4.5 Anwendung: Metriken berechnen

# Berechne die Demographic Parity Difference für das Test-Set
dp_diff  = demographic_parity_diff(df_test, 'is_black', 'y_pred')
# Berechne die False-Positive-Rate Difference für das Test-Set
fpr_diff = false_positive_rate_diff(df_test, 'is_black', 'y_true', 'y_pred')

# 4.5 Ausgabe: Ergebnisse anzeigen
print(f"Demographic Parity Difference:        {dp_diff:.3f}")
print(f"False Positive Rate Difference (FPR): {fpr_diff:.3f}")
```

### 4.6 Visualisierung

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# 1) Demographic Parity: Balkendiagramm der positiven Vorhersageraten

# a) Berechne die mittlere y_pred für jede Gruppe
rates = (
    df_test
    .groupby('is_black')['y_pred']
    .mean()
    .rename({0: 'Non-Black', 1: 'Black'})
)
# b) Zeichne das Balkendiagramm
plt.figure()
rates.plot(kind='bar')                                 # Balken-Chart
plt.ylabel('Predicted Recidivism Rate')                # Achsenbeschriftung
plt.title('Demographic Parity: Non-Black vs. Black')   # Diagrammtitel
plt.tight_layout()                                     # Layout anpassen
plt.show()                                             # Anzeige

# 2) False-Positive-Rate: Balkendiagramm der FPR je Gruppe

# a) Führe erneut pro Gruppe die Konfusionsmatrix durch und berechne FPR
fprs = []
labels = []
for g, sub in df_test.groupby('is_black'):
    tn, fp, fn, tp = confusion_matrix(
        sub['y_true'], sub['y_pred']
    ).ravel()                                          # Konfusionsmatrix
    fpr = fp / (fp + tn)                               # FP / (FP + TN)
    fprs.append(fpr)                                   # FPR speichern
    labels.append('Black' if g == 1 else 'Non-Black')  # Label für Achse

# b) Zeichne das Balkendiagramm
plt.figure()
plt.bar(labels, fprs)                                  # Balken-Chart
plt.ylabel('False Positive Rate')                      # Achsenbeschriftung
plt.title('False Positive Rate by Race')               # Diagrammtitel
plt.tight_layout()                                     # Layout anpassen
plt.show()                                             # Anzeige
```


### 4.6.1 Visualisierungs Erklärung

#### Chart 1

1. **Demographic Parity**  
   - **Was zeigt das Chart?**  
     Die Balken visualisieren die durchschnittliche „Rückfall“-Vorhersagerate (y_pred) für Nicht-Schwarze vs. Schwarze.  
   - **Interpretation:**  
     Liegt die Rate bei Schwarzen höher, so stuft das Modell diese Gruppe häufiger als „rückfällig“ ein, obwohl beide Gruppen gleich behandelt werden sollten. Die absolute Differenz der Balkenhöhen ist die **Demographic Parity Difference**.

2. **False Positive Rate (FPR)**  
   - **Was zeigt das Chart?**  
     Die Balken stehen für die Falsch-Positiv-Rate (FP / (FP + TN)) – also wie oft das Modell „Rückfall“ vorhersagt, obwohl keiner vorliegt – getrennt nach Gruppe.  
   - **Interpretation:**  
     Eine höhere FPR bei Schwarzen bedeutet, dass sie unverhältnismäßig oft fälschlich als „hochriskant“ eingestuft werden. Die Differenz der Balkenhöhen ist die **FPR Difference**.


- **Verletzung von Group Fairness:**         
  – **Demographic Parity** (ungleiche Vorhersageraten)  
  – **Equalized Odds** (ungleiche FPR)

- **Zutreffende Bias-Typen (von den sechs):**  
  - **Historical Bias:** Die Trainingsdaten spiegeln langjährige gesellschaftliche Ungleichheiten im Justizsystem wider und werden vom Modell übernommen.  
  - **Measurement Bias:** Die verwendeten Proxy-Variablen (z. B. COMPAS-Score, Vorstrafenzahl) sind für Schwarze und Nicht-Schwarze unterschiedlich akkurat, was sich in der höheren FPR für Schwarze zeigt.

- **Konsequenz:**  
  Eine systematische Überbestrafung Schwarzer 

- **Mögliche Gegenmaßnahmen:**  
  1. **Pre-Processing:** Daten ausgleichen (Reweighing, Sampling)  
  2. **In-Processing:** Fairness-Constraints im Training  
  3. **Post-Processing:** Schwellenwert-Anpassung zur FPR-Angleichung  

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# 1) Demographic Parity: Balkendiagramm der positiven Vorhersageraten

# a) Berechne die mittlere y_pred für jede Gruppe
rates = (
    df_test
    .groupby('is_black')['y_pred']
    .mean()
    .rename({0: 'Non-Black', 1: 'Black'})
)

fig, axes = plt.subplots(ncols=2, figsize=(12, 5))
display_labels = ['No Recidivism', 'Recidivism']
group_names = ['Non-Black', 'Black']

cms = []  # Store confusion matrices for colorbar scaling

for ax, (group_value, group_name) in zip(axes, enumerate(group_names)):
    subset = df_test[df_test['is_black'] == group_value]
    cm = confusion_matrix(subset['y_true'], subset['y_pred'], normalize='true')
    cms.append(cm)
    
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=display_labels)
    disp.plot(ax=ax, cmap=plt.cm.Blues, colorbar=False, values_format='.2f')
    ax.set_title(f'Normalized Confusion Matrix – {group_name}')
    ax.set_xlabel('Predicted label')
    ax.set_ylabel('True label')

# Eine Farbskala rechts hinzufügen, basierend auf dem Maximalwert aller Konfusionsmatrizen  
fig.colorbar(
    plt.cm.ScalarMappable(cmap=plt.cm.Blues, norm=plt.Normalize(vmin=0, vmax=max(cm.max() for cm in cms))),
    ax=axes,
    location='right',
    shrink=0.8,
    label='Rate'
)

# Im zweiten Subplot die X- und Y-Achsenbeschriftungen sowie die Tick-Labels entfernen  
axes[1].set_ylabel('')
axes[1].set_yticklabels([])
print('Ergebnisse:')
```

#### Chart 2: Konfusionsmatrizen



Die beiden nebeneinander dargestellten, normalisierten Konfusionsmatrizen zeigen auf Gruppenebene, wie gut unser Modell bei Schwarzen vs. Nicht-Schwarzen abschneidet.

|                    | Predicted: Kein Rückfall | Predicted: Rückfall |
|--------------------|--------------------------|---------------------|
| **True: Kein Rückfall** | True Negative (TN)       | False Positive (FP) |
| **True: Rückfall**     | False Negative (FN)      | True Positive (TP)  |

Da jede Zeile auf 1 normiert ist, lesen wir direkt **Raten** ab:

1. **True Negative Rate (TN-Rate)** – oben links  
   Anteil der korrekt als „kein Rückfall“ klassifizierten Fälle.  
   - Nicht-Schwarz: 0,91 (91 %)  
   - Schwarz:      0,56 (56 %)

2. **False Positive Rate (FPR)** – oben rechts  
   Anteil der fälschlich als „Rückfall“ klassifizierten „kein-Rückfall“-Fälle.  
   - Nicht-Schwarz: 0,09 (9 %)  
   - Schwarz:      0,44 (44 %)

3. **False Negative Rate (FN-Rate)** – unten links  
   Anteil der fälschlich als „kein Rückfall“ eingestuften Rückfaller.  
   - Nicht-Schwarz: 0,60 (60 %)  
   - Schwarz:      0,27 (27 %)

4. **True Positive Rate (TPR)** – unten rechts  
   Anteil der korrekt als „Rückfall“ erkannten Rückfaller.  
   - Nicht-Schwarz: 0,40 (40 %)  
   - Schwarz:      0,73 (73 %)

---

#### Erkenntnisse

- Die **False-Positive-Rate** bei Schwarzen (44 %) ist deutlich höher als bei Nicht-Schwarzen (9 %).  
  → Verletzung von **Equalized Odds** und Hinweis auf **Measurement Bias** (Proxy-Variablen sind für Schwarze weniger akkurat).

- Gleichzeitig ist die **True-Positive-Rate** bei Schwarzen (73 %) höher als bei Nicht-Schwarzen (40 %).  
  → Bestätigung, dass das Modell historische Ungleichheiten im Justizsystem übernimmt und verstärkt (**Historical Bias**).

- Insgesamt erfüllt das Modell weder **Demographic Parity** noch **Equalized Odds**, was in der Praxis zu systematischen Benachteiligungen führen kann.  


## 5. Fazit



<div style="background-color: #7690AC; padding: 10px; border-radius: 8px;">
  <p style="margin: 0; color: white; font-size: 16px; line-height: 1.5;">
    Bias lässt sich in KI niemals vollständig eliminieren, sondern nur kontinuierlich messen, verstehen und reduzieren. Unsere Analyse zeigt, dass Bias je nach Typ und Perspektive unterschiedlich auftritt, sei es in ungleichen Vorhersageraten oder variierenden Fehlerraten und somit die Ergebnisse maßgeblich beeinflusst.
Wie unser COMPAS-Beispiel belegt, können diese Verzerrungen zu systematischen Diskriminierungen führen, wenn bestimmte Gruppen überproportional als „hochriskant“ eingestuft werden. Daher ist es entscheidend, in jeder Phase der ML-Pipeline, von Datenerfassung über Modelltraining bis hin zum Deployment, systematisch zu prüfen, wo Bias entstehen kann und seine Auswirkungen nachzuvollziehen. Auf dieser Basis können gezielte Gegenmaßnahmen (Pre-, In- und Post-Processing) ergriffen werden, auch wenn ein Rest-Bias stets in jedem maschinellen Lernsystem verbleibt.  
  </p>
